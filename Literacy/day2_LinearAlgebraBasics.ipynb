```python
#Linear Algebra basics
#import numpy and pyplot for graphs
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vectors are multi dimentional arrays (represented often as tuples and can hold data of any dimension) 
# === PART 1: Vectors as Arrays ===
print("=== VECTORS ===")
v1 = np.array([2, 3])  # A vector in 2D space
v2 = np.array([1, 4])
```

```python

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
```

```python

# Vector addition
v_sum = v1 + v2
print(f"v1 + v2 = {v_sum}")
```

```python

# Scalar multiplication
v_scaled = 3 * v1
print(f"3 * v1 = {v_scaled}")
```

```python

# Dot product (CRITICAL for ML!)
dot_product = np.dot(v1, v2)
print(f"v1 · v2 = {dot_product}")
print(f"Calculation: (2*1) + (3*4) = {2*1 + 3*4}")
```

```python

# === PART 2: Matrices ===
print("\n=== MATRICES ===")
# Think of this as a transformation or weights in a neural network
matrix_A = np.array([[1, 2],
                     [3, 4]])

print(f"Matrix A:\n{matrix_A}")
```

```python

# Matrix-vector multiplication (what happens in neural networks!)
result = np.dot(matrix_A, v1)
print(f"\nMatrix A × Vector v1 = {result}")
print("This is what happens in EVERY layer of a neural network!")
```

```python

# === PART 3: Connect to Your NN Notes ===
print("\n=== NEURAL NETWORK CONNECTION ===")
# Remember: "weights assigned to each neural network node"
# Those weights are organized in MATRICES!

# Simple example: 2 inputs, 3 neurons in next layer
inputs = np.array([0.5, 0.8])  # Your input data
weights = np.array([[0.2, 0.4],   # Weights for neuron 1
                    [0.1, 0.9],   # Weights for neuron 2
                    [0.7, 0.3]])  # Weights for neuron 3
```

```python

# Each neuron computes: weighted sum of inputs
neuron_outputs = np.dot(weights, inputs)
print(f"Input: {inputs}")
print(f"Neuron outputs (before activation): {neuron_outputs}")
```

```python

# Add bias (remember from your notes: "biases for activity")
biases = np.array([0.1, 0.2, 0.15])
neuron_outputs_with_bias = neuron_outputs + biases
print(f"Neuron outputs (with bias): {neuron_outputs_with_bias}")
```

```python

# Apply ReLU activation (from your notes!)
relu_output = np.maximum(0, neuron_outputs_with_bias)
print(f"After ReLU activation: {relu_output}")
print("\nThis is literally how neural networks work!")

```
