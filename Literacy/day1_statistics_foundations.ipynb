```python
# Statistics Refresher - Day 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Generate sample data
data = np.random.normal(100, 15, 1000)  # mean=100, std=15, 1000 samples
```

```python

# Calculate manually
mean_manual = np.sum(data) / len(data)
variance_manual = np.sum((data - mean_manual)**2) / len(data)
std_manual = np.sqrt(variance_manual)
```

```python

# Verify with numpy
print(f"Manual Mean: {mean_manual:.2f}, NumPy Mean: {np.mean(data):.2f}")
print(f"Manual Std: {std_manual:.2f}, NumPy Std: {np.std(data):.2f}")
```

```python

# Visualize
plt.hist(data, bins=50, alpha=0.7)
plt.axvline(mean_manual, color='red', label=f'Mean: {mean_manual:.2f}')
plt.axvline(mean_manual + std_manual, color='orange', linestyle='--', label=f'+1 SD')
plt.axvline(mean_manual - std_manual, color='orange', linestyle='--', label=f'-1 SD')
plt.legend()
plt.show()
```

```python
# Linear Algebra Basics in NumPy
import numpy as np
```

```python

# Vectors (1D arrays)
v1 = np.array([2, 3])
v2 = np.array([1, 4])
```

```python

# Vector operations
print("Addition:", v1 + v2)
print("Scalar multiplication:", 3 * v1)
print("Dot product:", np.dot(v1, v2))  # This is CRITICAL for ML
```

```python

# Matrices (2D arrays)
matrix_A = np.array([[1, 2], [3, 4]])
matrix_B = np.array([[5, 6], [7, 8]])

# Matrix operations
print("Matrix multiplication:\n", np.matmul(matrix_A, matrix_B))

# WHY THIS MATTERS FOR ML:
# Neural networks are just matrix multiplications!
# Input * Weights = Output (simplified)
```
