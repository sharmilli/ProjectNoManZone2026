```python
# Create this notebook: understanding_matrices.ipynb

import numpy as np
```

```python

# Neural network weight matrix example
# 3 input features → 5 neurons in hidden layer
weights_input_to_hidden = np.random.randn(5, 3)  # 5×3 matrix
```

```python

print(f"Weight matrix shape: {weights_input_to_hidden.shape}")
print(f"This means: {weights_input_to_hidden.shape[0]} neurons, each receiving {weights_input_to_hidden.shape[1]} inputs")
```

```python

# Sample input (3 features)
input_data = np.array([1.2, 0.8, 2.1])
```

```python

# Compute hidden layer activations
hidden_layer = np.dot(weights_input_to_hidden, input_data)
```

```python
print(f"\nInput shape: {input_data.shape}")
print(f"Hidden layer output shape: {hidden_layer.shape}")
print(f"We went from {len(input_data)} inputs to {len(hidden_layer)} neurons")

# Question to answer in your notes:
# If you have 10 input features and want 20 neurons in the next layer,
# what should be the shape of your weight matrix?
```

```python

weights_input_to_hidden = np.random.randn(20,10)  # 5×3 matrix

print(f"Weight matrix shape: {weights_input_to_hidden.shape}")
print(f"This means: {weights_input_to_hidden.shape[0]} neurons, each receiving {weights_input_to_hidden.shape[1]} inputs")
```

```python

# Sample input (20 features)
input_data = np.array([1.1,1.2,1.3,1.4,1.5,1.6,1.7,1.8,1.9,1.10])

# Compute hidden layer activations
hidden_layer = np.dot(weights_input_to_hidden, input_data)
print(f"\nInput shape: {input_data.shape}")
print(f"Hidden layer output shape: {hidden_layer.shape}")
print(f"We went from {len(input_data)} inputs to {len(hidden_layer)} neurons")
```
